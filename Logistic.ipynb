```python
import numpy as np
from sklearn.metrics import *
import matplotlib.pyplot as plt
```

```python
# Create Data X and Y
x = np.arange(100)
y = np.array([2*i + np.random.randint(-5,5) for i in x])
plt.scatter(x,y)
```

```python
a = 100 # a = 2
lr = 0.0001
epochs = 20
N = len(y)
# Training
for i in range(10):
    # Step 1: Calculate gradient at a
    y_pred = a*x
    dL_da = (2/N)*np.sum((y_pred-y)*x)
    # Print Loss
    L = (1/N)*np.sum((y_pred-y)**2)
    print(f'Epochs {i}: \n\t MSE: {L}',end='\t')
    print(f'a = {a}')
    # Gradient descent
    a = a-dL_da*lr
plt.plot(y_pred,c = 'r')
```
